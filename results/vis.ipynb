```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(12,8))
x = [99.9, 99, 97, 95, 93, 90, 85, 80]

y = np.array([[77.22333333, 80.70684792, 73.33333333],
              [75.84333333, 79.77301483, 80.66666667],
              [71.39333333, 75.63254981, 85.33333333],
              [69.88333333, 75.02372857, 86.66666667],
              [67.20666667, 70.62246425, 89.],
              [67.20666667, 70.62246425, 89.],
              [65.20666667, 68.34985834, 91.66666667],
              [65.20666667, 68.34985834, 91.66666667]])

font = font_manager.FontProperties(family='Times New Roman',
                                   weight='bold',
                                   style='normal',
                                   size=21)

makers = ['^', '*', 'o']
colors = ['yellow', 'red', 'blue']
linestyles = ['-','--','-.']
metrics = ['AUROC', 'AUPR', 'FPR']
# Plot the curves
for i in range(y.shape[1]):
    plt.plot(x, y[:, i], marker =makers[i], linestyle=linestyles[i], linewidth=3, markersize=10)

# Set the x-axis and y-axis limits
plt.xlim(100, 80)
plt.ylim(65, 95)

# Set the x-axis labels
plt.xticks(x,rotation=45)
plt.tick_params(labelsize=20)
# Add labels and title
plt.legend(['AUROC','AUPR','FPR'],prop=font)
plt.xlabel('the cumulative explained variance in percentage',fontfamily='Times New Roman',fontsize=30,fontweight='bold')
plt.ylabel('the average metric in percentage',fontfamily='Times New Roman',fontsize=30,fontweight='bold')
plt.grid()
plt.savefig('PCA.png', dpi=300, bbox_inches='tight')
# Show the plot
plt.show()
```
